# Wrapper methods with Kfold cross validation

```python
import pandas as pd
import LogisticRegression as LR
import numpy as np
```

```python
from sklearn import datasets
```

```python
data = datasets.load_wine()
```

```python
Data = pd.concat([pd.DataFrame(data.data, columns = data.feature_names), pd.Series(data.target, name = 'target')], axis=1)
```

```python
Data
```

```python
class Kfold:
    def __init__(self, K=5):
        self.K = K
    def split(self, data):
        Random = list(range(len(data)))
        np.random.shuffle(Random)
        data = data.iloc[Random]
        Folds = []
        n = len(data)
        for i in range(self.K):
            Folds.append(data.iloc[i*(n//self.K) : (i+1)*(n//self.K)])
        return Folds
    def evaluate_model(self, model, data):
        Folds = self.split(data)
        Error = []
        for i in range(self.K):
            Test_data = Folds[i]
            Train_data = pd.DataFrame([], columns = Test_data.columns)
            for j in range(self.K):
                if j != i:
                    Train_data = pd.concat([Train_data, Folds[j]])
            model.fit(Train_data.iloc[:,:-1], Train_data.iloc[:,-1])
            Error.append((model.predict(Test_data.iloc[:,:-1].values)==Test_data.iloc[:,-1].values).mean())
        return Error
```

```python
model = LR.MulticlassLogReg(lr=0.08, Max_iter=200)
KFold = Kfold()
```

```python
model.fit(Data.iloc[:,:-1], Data.iloc[:,-1])
```

```python
model.predict(Data.iloc[:,:-1].values)
```

```python
(model.predict(Data.iloc[:,:-1].values)==Data.iloc[:,-1].values).mean()
```

```python
KFold.evaluate_model(model=model, data=Data)
```
